```python
sb.countplot(x = df.Department)
plt.xlabel("Departments")
plt.ylabel("Number Of Employees")
plt.title("Distribution of employees across departments")
plt.xticks(rotation = 90)
plt.show()
```

1. Maximum employees are under Research and Development department.
2. The Human Resources department has lowest number of employees.

4. This graph states that company is majorly focused on Research & Development department.

```python
sb.countplot(x = df.JobRole)
plt.xlabel("Job Roles")
plt.ylabel("Number Of Employees")
plt.title("Distribution of employees across Job Roles")
plt.xticks(rotation = 90)
plt.show()
```

Distribution of Employees across Job Roles insights : 

1. The sales executive role has the highest number of employees around 326.
2. The Sales Representative, Resource Director and Human resources role has the least number of employees and are small job roles.
3. This states that the company is majorly focused on Sales executive, Research Scientist and Laboratory Technician role.
4. Manufacturing Director, Healthcare Representative and Manager is mid-sized role.

```python
sb.histplot(x = df.Age, kde = True, color = "blue")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Distribution of Age")
plt.show
```

Insights from the Age Distribution Histogram : 

1. Most of the employees are from age group 25 to 40 which means that company is having young to mid-career employees that indicates innovation and        energy.

```python
sb.histplot(x=df.MonthlyIncome, kde=True, color="green")
plt.xlabel("Monthly Income")
plt.ylabel("Frequency")
plt.title("Distribution of Monthly Income")
plt.show()
```

Insights from the Monthly Income Histogram :

1. The histogram shows a high number of employees earning between 2000 - 5000 per month.
2. Salaries gradually decreases when we move towards higher income level.
3. The number of employees earning more than 10000 per month are very less which indicates that there are limited higher paying positions.
4. The largest group of employees earn around 2500 per month.

```python
corr=df.corr(numeric_only=True)
corr
```

```python
plt.figure(figsize=(20,11))
sb.heatmap(corr,annot=True)
plt.show()
```

1. Strong Correlation (Closer to 1 or -1)
   - Job Level vs Monthly Income(0.95) : Higher job level are strongly correlated with higher salaries.
   - Total Working Years vs Job Level(0.78) : Employees with more working year are having higher job level.
   - Total Working Years vs Monthly Income(0.77) : Employees with more working year tends to have higher income.
   - Performance Rating vs Percent Salary Hike(0.77) : Employees with higher performance rating tends to have higher and faster salary hike.
   - Years At Company vs Years With Current Manager(0.77) : If an employee has been at the company for a long time, they are likely to have worked under the same manager for a significant portion of that time. We can also see that as the correlation is stronger the employees stick with the same manager for years.
   - Years at Company vs Years in Current Role(0.76) : As the correlation is strong we can say that the roles of employees are not frequently changed and they work under the same role for longer period of time, but it may lead to limited growth opportunities.

2. Weak Correlation (< 0.3)
   - Distance From Home vs Monthly Income(-0.017) : The distance an employees lives from the office has almost no impact on their salary.
   - Education vs Job Satisfaction(-0.011) : Higher education does not lead to higher job satisfaction.
   - Age vs Distance from Home(-0.0017) : Age and distance from home are completely unrelated.
   - Performance Rating vs Monthly Income(-0.017) : Performance rating does not have any strong impact on salary.

3. Moderate Correlation(0.5 - 0.7)
    - Years at Company vs Years Since Last Promotion(0.62) : Employees who have spent more years in the company tend to have more promotions.
    - Years in Current Role vs Years Since Last Promotion(0.55) : Employees who stay in the same role for long tend to have fewer promotions.
   
   

```python
attrition = df.Attrition.value_counts()
attrition
```

```python
sb.countplot(x = df.Attrition)
plt.xlabel("Attrition Status")
plt.ylabel("Number Of Employees")
plt.title("Distribution of employees across Job Roles")
plt.xticks(rotation = 90)
plt.show()
```

Distribution of Employees Across Job Roles Insights :

- The above countplot shows that around 237 employees have left the company where as around 1233 employees remain with the company.
- This suggests that most employees remain with the company, while only a small percentage leave.

```python
sb.boxplot(x = df.Attrition, y = df.JobSatisfaction)
plt.xlabel("Attrition Status")
plt.ylabel("Job Satisfaction")
plt.title("Distribution of Job Satisfaction Across Attrition Status")
plt.xticks(rotation = 90)
plt.show()
```

Distribution of Job Satisfaction Across Attrition Status Insights : 

- The boxplot states that employees who left had lower job satisfaction where as employees who stayed had higher job satisfaction on average.

```python
sb.boxplot(x = df.Attrition, y = df.MonthlyIncome)
plt.xlabel("Attrition Status")
plt.ylabel("Monthly Income")
plt.title("Distribution of Monthly Income Across Attrition Status")
plt.xticks(rotation = 90)
plt.show()
```

Distributin of Monthly Income Across Attrition Status Insights :

- The boxplot states that the employees who left the company had low monthly income than the ones who stayed.
- There are extreme salary outliers in both categories, especially amoung employees who stayed, but some employees with very high salaries(>20,000) still left, indicating that salary alone is not the only factor influencing attrition.
- So we can conclude that increasing pay for low income employees might reduce attrition also some high income employees still left the company so company must also improve job satisfaction, work life balance and career growth opportunities.  

```python
sb.boxplot(x = df.Attrition, y = df.PerformanceRating)
plt.xlabel("Attrition Status")
plt.ylabel("Performance Rating")
plt.title("Distribution of Performance Rating Across Attrition Status")
plt.xticks(rotation = 90)
plt.show()
```

Distribution of Performance Rating Across Attrition Status Insights : 

1. Almost all performance ratings are concentrated at 3, regardless of whether employees left or stayed.
2. Since the distribution is nearly identical for both attrition groups, performance rating does not seem to significantly influence attrition in this dataset.

```python
df.groupby('Attrition')['PerformanceRating'].value_counts()
```

Distribution of Performance Rating Across Attrition Status Insights :

- Most of the employees(both those who stayes and those who left) have a performance rating of 3.
- Also we can say that the employees with a higher rating(4) were less likely to leave, possibly due to better rewards, promotions, or job satisfaction.
- Since most employees have a rating of 3, performance rating alone may not be a strong reason of attrition.

```python
sb.boxplot(x = df.Attrition, y = df.Age)
plt.xlabel("Attrition Status")
plt.ylabel("Age")
plt.title("Distribution of Age Across Attrition Status")
plt.xticks(rotation = 90)
plt.show()
```

Distribution of Age Across Attrition Status Insights :

- The employees who left the company where younger as compared to the employees who stayed.
- Older employees might be more settled in their careers and less likely to switch jobs.
- Also there are few outliers in the "Yes" category, as some employees over 50 have left. It may be due to retirement, job dissatisfaction, or better opportunities elsewhere.
